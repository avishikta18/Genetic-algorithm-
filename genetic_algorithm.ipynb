```python
import numpy as np

class Genetic(object):

    def __init__(self, f, pop_size=100, n_variables=2):
        self.f = f
        self.minim = 1
        self.maxim = 10
        self.pop_size = pop_size
        self.n_variables = n_variables
        self.population = self.initializePopulation()
        self.evaluatePopulation()

    def initializePopulation(self):
        # Initialize population with random values for x, then calculate y as y = 10 - x
        return [np.random.randint(self.minim, self.maxim, size=(1))[0] for _ in range(self.pop_size)]

    def evaluatePopulation(self):
        # Evaluate the fitness of the population based on the function, using x and y where y = 10 - x
        return [self.f(i, 10 - i) for i in self.population]

    def nextGen(self):
        results = self.evaluatePopulation()
        # Select the best individual initially (for maximization, choose the individual with the highest fitness value)
        children = [self.population[np.argmax(results)]]  # Select the individual with the maximum fitness

        while len(children) < self.pop_size:
            # Tournament selection
            randA, randB = np.random.randint(0, self.pop_size), np.random.randint(0, self.pop_size)
            if results[randA] > results[randB]:  # Maximization: select the individual with the higher fitness
                p1 = self.population[randA]
            else:
                p1 = self.population[randB]

            randA, randB = np.random.randint(0, self.pop_size), np.random.randint(0, self.pop_size)
            if results[randA] > results[randB]:  # Maximization: select the individual with the higher fitness
                p2 = self.population[randA]
            else:
                p2 = self.population[randB]

            # Crossover (x1, y1) with (x2, y2)
            # Ensure after crossover the sum remains 10 (x + y = 10)
            p1_x, p1_y = p1, 10 - p1
            p2_x, p2_y = p2, 10 - p2

            # Recombination: Swap genes (crossover) and make sure x + y = 10
            # Since the sum must remain 10, if we swap x, we will adjust y accordingly
            child_x = (p1_x + p2_x) // 2
            child_y = 10 - child_x  # Ensure x + y = 10

            # Add child to the new population
            children.append(child_x)

        self.population = children

    def run(self):
        ix = 0
        while ix < 1000:
            ix += 1
            self.nextGen()
        return self.population[0]  # Return the best solution found

# Function to maximize (objective function)
f = lambda x, y: (x-6)**2 + y**2  # f(x, y) = (x-6)^2 + y^2

# Create the Genetic Algorithm instance
gen = Genetic(f)

# Run the algorithm and find the maximum (x, y)
maxim = gen.run()

# Calculate y as 10 - x to satisfy the condition
maxim_y = 10 - maxim

# Print the maximum values of x and y
print('Maximum found      X =', maxim, ', Y =', maxim_y)

# Calculate the function value at the maximum (x, y)
maxim_value = f(maxim, maxim_y)
print('Function value at maximum X, Y =', maxim_value)
```

```python
import numpy as np
import random

# Rosenbrock function definition
def rosenbrock(x):
    return 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2

# Genetic Algorithm parameters
population_size = 100
num_generations = 500
mutation_rate = 0.1
crossover_rate = 0.7
num_variables = 2  # 2 variables for Rosenbrock function

# Initialize the population
def create_population(population_size, num_variables):
    population = []
    for _ in range(population_size):
        individual = [random.uniform(-5, 5) for _ in range(num_variables)]
        population.append(individual)
    return population

# Tournament selection
def tournament_selection(population, fitness_values, tournament_size=5):
    selected = random.sample(list(zip(population, fitness_values)), tournament_size)
    selected = sorted(selected, key=lambda x: x[1])  # Sort by fitness (minimize)
    return selected[0][0]

# Crossover (Single-point crossover)
def crossover(parent1, parent2):
    if random.random() < crossover_rate:
        crossover_point = random.randint(1, num_variables-1)
        child1 = parent1[:crossover_point] + parent2[crossover_point:]
        child2 = parent2[:crossover_point] + parent1[crossover_point:]
        return child1, child2
    else:
        return parent1, parent2

# Mutation
def mutate(individual):
    if random.random() < mutation_rate:
        mutation_point = random.randint(0, num_variables-1)
        individual[mutation_point] += random.uniform(-1, 1)
    return individual

# Main Genetic Algorithm loop
def genetic_algorithm():
    # Create initial population
    population = create_population(population_size, num_variables)

    # Store the best solution
    best_solution = None
    best_fitness = float('inf')

    # Store the generations and fitness values
    generations_fitness = []

    for generation in range(num_generations):
        # Evaluate fitness for each individual
        fitness_values = [rosenbrock(individual) for individual in population]

        # Update the best solution
        min_fitness = min(fitness_values)
        min_index = fitness_values.index(min_fitness)
        if min_fitness < best_fitness:
            best_fitness = min_fitness
            best_solution = population[min_index]

        # Record generation and best fitness
        generations_fitness.append((generation, best_fitness))

        # Create new population via selection, crossover, and mutation
        new_population = []
        for _ in range(population_size // 2):
            parent1 = tournament_selection(population, fitness_values)
            parent2 = tournament_selection(population, fitness_values)

            # Crossover to create children
            child1, child2 = crossover(parent1, parent2)

            # Mutate children
            child1 = mutate(child1)
            child2 = mutate(child2)

            # Add children to the new population
            new_population.append(child1)
            new_population.append(child2)

        population = new_population

    return best_solution, best_fitness, generations_fitness

# Run the algorithm
best_solution, best_fitness, generations_fitness = genetic_algorithm()

# Print the best solution and fitness
print(f"Best Solution: {best_solution}")
print(f"Best Fitness: {best_fitness}")

# Print generations and their fitness values
print("\nGenerations and their Best Fitness:")
for generation, fitness in generations_fitness:
    print(f"Generation {generation}: Best Fitness = {fitness}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Rastrigin function
def rastrigin(X):
    A = 10
    return A * len(X) + np.sum(X**2 - A * np.cos(2 * np.pi * X))

# Genetic Algorithm Parameters
POPULATION_SIZE = 100
NUM_GENERATIONS = 1000
DIMENSIONS = 10  # Number of variables in Rastrigin function
BOUND_LOW = -5.12
BOUND_UP = 5.12
CROSSOVER_RATE = 0.8
MUTATION_RATE = 0.1
TOURNAMENT_SIZE = 3
ELITISM = True

# Initialize population
def initialize_population(size, dimensions, lower_bound, upper_bound):
    return np.random.uniform(low=lower_bound, high=upper_bound, size=(size, dimensions))

# Evaluate fitness (inverse of Rastrigin function value)
def evaluate_fitness(population):
    fitness = np.array([1 / (1 + rastrigin(individual)) for individual in population])
    return fitness

# Tournament Selection
def tournament_selection(population, fitness, tournament_size):
    selected_indices = []
    for _ in range(len(population)):
        participants = np.random.choice(len(population), tournament_size, replace=False)
        winner = participants[np.argmax(fitness[participants])]
        selected_indices.append(winner)
    return population[selected_indices]

# Crossover (Uniform Crossover)
def crossover(parent1, parent2, crossover_rate):
    if np.random.rand() < crossover_rate:
        mask = np.random.randint(0, 2, size=parent1.shape).astype(bool)
        child1 = np.where(mask, parent1, parent2)
        child2 = np.where(mask, parent2, parent1)
        return child1, child2
    else:
        return parent1.copy(), parent2.copy()

# Mutation (Gaussian Mutation)
def mutate(individual, mutation_rate, lower_bound, upper_bound):
    for i in range(len(individual)):
        if np.random.rand() < mutation_rate:
            individual[i] += np.random.normal()
            individual[i] = np.clip(individual[i], lower_bound, upper_bound)
    return individual

# Genetic Algorithm
def genetic_algorithm():
    population = initialize_population(POPULATION_SIZE, DIMENSIONS, BOUND_LOW, BOUND_UP)
    best_fitness_over_time = []
    best_individual = None
    best_fitness = -np.inf

    for generation in range(NUM_GENERATIONS):
        fitness = evaluate_fitness(population)

        max_fitness = np.max(fitness)
        if max_fitness > best_fitness:
            best_fitness = max_fitness
            best_individual = population[np.argmax(fitness)].copy()

        best_fitness_over_time.append(1 / best_fitness - 1)  # Convert back to Rastrigin value

        selected_population = tournament_selection(population, fitness, TOURNAMENT_SIZE)
        next_generation = []
        np.random.shuffle(selected_population)

        for i in range(0, POPULATION_SIZE, 2):
            parent1 = selected_population[i]
            if i+1 < POPULATION_SIZE:
                parent2 = selected_population[i+1]
            else:
                parent2 = selected_population[0]

            child1, child2 = crossover(parent1, parent2, CROSSOVER_RATE)
            child1 = mutate(child1, MUTATION_RATE, BOUND_LOW, BOUND_UP)
            child2 = mutate(child2, MUTATION_RATE, BOUND_LOW, BOUND_UP)

            next_generation.extend([child1, child2])

        population = np.array(next_generation[:POPULATION_SIZE])

        if ELITISM:
            worst_index = np.argmin(fitness)
            population[worst_index] = best_individual.copy()

        if (generation+1) % 100 == 0 or generation == 0:
            print(f"Generation {generation+1}: Best Rastrigin Value = {1 / best_fitness - 1:.5f}")

    print("\nOptimization Complete!")
    print(f"Best Rastrigin Value Found: {1 / best_fitness - 1:.5f}")
    print(f"Best Individual: {best_individual}")



if __name__ == "__main__":
    genetic_algorithm()
```

```python
import numpy as np

# Six-Hump Camel Back Function
def six_hump_camel_back(x):
    x1, x2 = x[0], x[1]
    term1 = (4 - 2.1 * x1**2 + (x1**4) / 3) * x1**2
    term2 = x1 * x2
    term3 = (-4 + 4 * x2**2) * x2**2
    return term1 + term2 + term3

# Genetic Algorithm Parameters
POP_SIZE = 100  # Population size
GENS = 100  # Number of generations

# Initialize Population
def initialize_population():
    return np.random.uniform(-3, 3, (POP_SIZE, 2))

# Selection
def select_parents(population, fitness):
    # Select parents based on fitness (using tournament selection, for example)
    idx = np.argsort(fitness)
    return population[idx[:POP_SIZE//2]]

# Crossover (Single Point Crossover)
def crossover(parents):
    offspring = []
    for i in range(0, len(parents), 2):
        parent1, parent2 = parents[i], parents[i + 1]
        crossover_point = np.random.randint(2)
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        offspring.append(child1)
        offspring.append(child2)
    return np.array(offspring)

# Mutation
def mutate(offspring, mutation_rate=0.1):
    for i in range(len(offspring)):  # Ensure mutation operates within the bounds of the offspring
        if np.random.rand() < mutation_rate:
            mutation = np.random.normal(0, 0.1, 2)
            offspring[i] += mutation
    return offspring

# Elitism (Preserving the best individual)
def elitism(population, fitness, offspring):
    best_idx = np.argmin(fitness)
    best_individual = population[best_idx]
    # Keep the best individual
    offspring[0] = best_individual
    return offspring

# Main Genetic Algorithm
def genetic_algorithm():
    population = initialize_population()
    best_solution = None
    best_fitness = float('inf')
    generation_results = []  # To track best fitness and solution for all generations

    for gen in range(GENS):
        # Calculate fitness for each individual
        fitness = np.apply_along_axis(six_hump_camel_back, 1, population)

        # Track the best solution
        gen_best = np.min(fitness)
        if gen_best < best_fitness:
            best_fitness = gen_best
            best_solution = population[np.argmin(fitness)]

        generation_results.append((gen, best_solution, best_fitness))  # Store results of each generation

        # Selection
        parents = select_parents(population, fitness)

        # Crossover
        offspring = crossover(parents)

        # Mutation
        offspring = mutate(offspring)

        # Elitism: Keep the best individual
        population = elitism(population, fitness, offspring)

    return generation_results

# Run the algorithm and get the results for all generations
generation_results = genetic_algorithm()

# Print results of all generations
for gen, best_solution, best_fitness in generation_results:
    print(f"Generation {gen}: Best Solution = {best_solution}, Best Fitness = {best_fitness}")
    # Run the algorithm and get the results for all generations
generation_results = genetic_algorithm()

# Print results of all generations
for gen, best_solution, best_fitness in generation_results:
    print(f"Generation {gen}: Best Solution = {best_solution}, Best Fitness = {best_fitness}")

# Display the final best fitness
final_gen, final_solution, final_fitness = generation_results[-1]
print(f"\nFinal Best Fitness: {final_fitness}")
print(f"Final Best Solution: {final_solution}")

```

```python
import numpy as np

# Ackley Function
def ackley_function(x, a=20, b=0.2, c=2 * np.pi):
    d = len(x)
    term1 = -a * np.exp(-b * np.sqrt(np.sum(x**2) / d))
    term2 = -np.exp(np.sum(np.cos(c * x)) / d)
    return term1 + term2 + a + np.exp(1)

# Genetic Algorithm Parameters
POP_SIZE = 100  # Population size
GENS = 100  # Number of generations
DIMENSIONS = 2  # Dimensions of the input
MUTATION_RATE = 0.1  # Mutation rate
LOWER_BOUND = -32.768  # Lower bound of the search space
UPPER_BOUND = 32.768  # Upper bound of the search space

# Initialize Population
def initialize_population():
    return np.random.uniform(LOWER_BOUND, UPPER_BOUND, (POP_SIZE, DIMENSIONS))

# Selection: Tournament Selection
def select_parents(population, fitness):
    selected = []
    for _ in range(len(population) // 2):
        i, j = np.random.randint(0, len(fitness), 2)
        if fitness[i] < fitness[j]:
            selected.append(population[i])
        else:
            selected.append(population[j])
    # Ensure the number of parents is even
    if len(selected) % 2 != 0:
        selected = selected[:-1]
    return np.array(selected)

# Crossover: Single Point Crossover
def crossover(parents):
    offspring = []
    if len(parents) < 2:  # Ensure at least two parents for crossover
        return np.array(offspring)
    for i in range(0, len(parents), 2):
        if i + 1 >= len(parents):
            break  # Skip if there are no pairs left
        parent1, parent2 = parents[i], parents[i + 1]
        crossover_point = np.random.randint(1, DIMENSIONS)  # Random point in [1, DIMENSIONS-1]
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        offspring.append(child1)
        offspring.append(child2)
    return np.array(offspring)

# Mutation: Gaussian Mutation
def mutate(offspring):
    for i in range(len(offspring)):
        if np.random.rand() < MUTATION_RATE:
            mutation = np.random.normal(0, 1, DIMENSIONS)
            offspring[i] += mutation
            offspring[i] = np.clip(offspring[i], LOWER_BOUND, UPPER_BOUND)
    return offspring

# Elitism: Preserve the best solution
def elitism(population, fitness, offspring):
    if len(offspring) == 0:  # Handle empty offspring
        return population
    best_idx = np.argmin(fitness)
    best_individual = population[best_idx]
    offspring[0] = best_individual  # Replace the first individual with the best
    return offspring

# Main Genetic Algorithm
def genetic_algorithm():
    population = initialize_population()
    best_solution = None
    best_fitness = float('inf')

    for gen in range(GENS):
        # Calculate fitness
        fitness = np.apply_along_axis(ackley_function, 1, population)

        # Track the best solution
        gen_best = np.min(fitness)
        if gen_best < best_fitness:
            best_fitness = gen_best
            best_solution = population[np.argmin(fitness)]

        print(f"Generation {gen}: Best Fitness = {gen_best}, Best Solution = {best_solution}")

        # Selection
        parents = select_parents(population, fitness)

        # Crossover
        offspring = crossover(parents)

        # Mutation
        offspring = mutate(offspring)

        # Handle cases where offspring is empty
        if len(offspring) == 0:
            offspring = initialize_population()

        # Elitism
        population = elitism(population, fitness, offspring)

    return best_solution, best_fitness

# Run the Genetic Algorithm
best_solution, best_fitness = genetic_algorithm()
print("\nFinal Best Solution:", best_solution)
print("Final Best Fitness:", best_fitness)
```

```python
import numpy as np

# Egg Holder function
def egg_holder(x):
    x1, x2 = x[0], x[1]
    return -(x2 + 47) * np.sin(np.sqrt(abs(x2 + x1 / 2 + 47))) - x1 * np.sin(np.sqrt(abs(x1 - (x2 + 47))))

# Initialize population
def initialize_population(pop_size, bounds):
    return np.random.uniform(bounds[0], bounds[1], (pop_size, len(bounds[0])))

# Evaluate fitness
def evaluate_fitness(population):
    return np.array([egg_holder(ind) for ind in population])

# Tournament selection
def tournament_selection(population, fitness, tournament_size=3):
    selected = []
    for _ in range(len(population)):
        indices = np.random.choice(len(population), tournament_size, replace=False)
        best = indices[np.argmax(fitness[indices])]
        selected.append(population[best])
    return np.array(selected)

# Crossover: Uniform Crossover
def uniform_crossover(parent1, parent2):
    mask = np.random.rand(len(parent1)) < 0.5
    child = np.where(mask, parent1, parent2)
    return child

# Mutation with adaptive rate
def mutate(individual, bounds, mutation_rate):
    if np.random.rand() < mutation_rate:
        idx = np.random.randint(len(individual))
        individual[idx] += np.random.uniform(-1, 1)
        individual[idx] = np.clip(individual[idx], bounds[0][idx], bounds[1][idx])
    return individual

# Genetic Algorithm
def genetic_algorithm(bounds, pop_size=50, generations=100, crossover_rate=0.8, mutation_rate=0.2, tol=1e-6):
    population = initialize_population(pop_size, bounds)
    best_solution, best_fitness = None, float('inf')
    no_improvement_count = 0

    for gen in range(1, generations + 1):
        fitness = evaluate_fitness(population)
        if np.min(fitness) < best_fitness:
            best_fitness = np.min(fitness)
            best_solution = population[np.argmin(fitness)]
            no_improvement_count = 0
        else:
            no_improvement_count += 1

        # Early stopping if no improvement for 10 generations
        if no_improvement_count >= 10:
            print(f"Stopping early at generation {gen}. No improvement for 10 generations.")
            break

        # Selection
        selected = tournament_selection(population, fitness)

        # Create next generation
        next_population = []
        for i in range(0, pop_size, 2):
            parent1, parent2 = selected[np.random.randint(pop_size)], selected[np.random.randint(pop_size)]
            if np.random.rand() < crossover_rate:
                child1 = uniform_crossover(parent1, parent2)
                child2 = uniform_crossover(parent2, parent1)
            else:
                child1, child2 = parent1, parent2

            next_population.append(mutate(child1, bounds, mutation_rate))
            next_population.append(mutate(child2, bounds, mutation_rate))

        population = np.array(next_population)

        # Log generation progress
        print(f"Generation {gen}, Best Fitness: {best_fitness}")

    return best_solution, best_fitness

# Run the Genetic Algorithm
bounds = [(-512, -512), (512, 512)]  # Egg Holder function bounds
best_sol, best_fit = genetic_algorithm(bounds)

print(f"Best Solution: {best_sol}, Best Fitness: {best_fit}")
```

```python
import numpy as np
import random

# Define Bohachevsky Functions
def bohachevsky_function_1(x1, x2):
    return x1**2 + 2 * x2**2 - 0.3 * np.cos(3 * np.pi * x1) - 0.4 * np.cos(4 * np.pi * x2) + 0.7

def bohachevsky_function_2(x1, x2):
    return x1**2 + 2 * x2**2 - 0.3 * np.cos(3 * np.pi * x1) * np.cos(4 * np.pi * x2) + 0.3

def bohachevsky_function_3(x1, x2):
    return x1**2 + 2 * x2**2 - 0.3 * np.cos(3 * np.pi * x1 + 4 * np.pi * x2) + 0.3

# Initialize population
def initialize_population(size, bounds):
    return np.random.uniform(bounds[0], bounds[1], (size, 2))

# Fitness calculation
def calculate_fitness(population, func):
    return np.array([func(ind[0], ind[1]) for ind in population])

# Selection
def selection(population, fitness, num_parents):
    parents_indices = np.argsort(fitness)[:num_parents]
    return population[parents_indices]

# Crossover
def crossover(parents, offspring_size):
    offspring = []
    for _ in range(offspring_size):
        p1, p2 = random.sample(range(len(parents)), 2)
        child = np.array([parents[p1][0], parents[p2][1]])
        offspring.append(child)
    return np.array(offspring)

# Mutation
def mutation(offspring, mutation_rate, bounds):
    for child in offspring:
        if random.random() < mutation_rate:
            gene_idx = random.randint(0, 1)
            child[gene_idx] = np.random.uniform(bounds[0], bounds[1])
    return offspring

# Genetic Algorithm
def genetic_algorithm(func, pop_size, generations, bounds, mutation_rate, num_parents):
    # Initialize population
    population = initialize_population(pop_size, bounds)
    best_solution = None
    best_fitness = float('inf')

    for generation in range(generations):
        # Calculate fitness
        fitness = calculate_fitness(population, func)

        # Update best solution
        min_idx = np.argmin(fitness)
        if fitness[min_idx] < best_fitness:
            best_fitness = fitness[min_idx]
            best_solution = population[min_idx]

        # Selection
        parents = selection(population, fitness, num_parents)

        # Crossover
        offspring_size = pop_size - num_parents
        offspring = crossover(parents, offspring_size)

        # Mutation
        offspring = mutation(offspring, mutation_rate, bounds)

        # Update population
        population = np.vstack((parents, offspring))

        # Print progress
        print(f"Generation {generation + 1}: Best Fitness = {best_fitness:.4f}")

    return best_solution, best_fitness

# Parameters
population_size = 50
num_generations = 100
bounds = [-100, 100]
mutation_rate = 0.1
num_parents = 10

# Select the function to optimize
functions = {
    "1": bohachevsky_function_1,
    "2": bohachevsky_function_2,
    "3": bohachevsky_function_3,
}

print("Select a Bohachevsky Function to Optimize:")
print("1: Bohachevsky Function 1")
print("2: Bohachevsky Function 2")
print("3: Bohachevsky Function 3")
choice = input("Enter your choice (1/2/3): ")

selected_function = functions.get(choice, bohachevsky_function_1)

# Run Genetic Algorithm
best_solution, best_fitness = genetic_algorithm(selected_function, population_size, num_generations, bounds, mutation_rate, num_parents)

# Output Results
print(f"\nOptimized for Bohachevsky Function {choice}")
print(f"Best Solution: {best_solution}")
print(f"Best Fitness: {best_fitness:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.optimize import minimize

# Define the objective function
def objective_function(x):
    return np.sum(x**2)  # Example: Minimize the sum of squares

# Initialize population
def initialize_population(size, dimensions, bounds):
    return np.random.uniform(bounds[0], bounds[1], (size, dimensions))

# Evaluate fitness
def evaluate_fitness(population):
    return np.array([objective_function(ind) for ind in population])

# Selection (Tournament)
def tournament_selection(population, fitness, tournament_size):
    selected_indices = []
    for _ in range(len(population)):
        participants = np.random.choice(len(population), tournament_size, replace=False)
        winner = participants[np.argmin(fitness[participants])]
        selected_indices.append(winner)
    return population[selected_indices]

# Crossover (Uniform)
def uniform_crossover(parent1, parent2, crossover_rate):
    if np.random.rand() < crossover_rate:
        mask = np.random.rand(len(parent1)) < 0.5
        child1, child2 = parent1.copy(), parent2.copy()
        child1[mask], child2[mask] = parent2[mask], parent1[mask]
        return child1, child2
    return parent1.copy(), parent2.copy()

# Mutation (Dynamic)
def mutation(individual, mutation_rate, bounds):
    if np.random.rand() < mutation_rate:
        mutation_index = np.random.randint(len(individual))
        individual[mutation_index] = np.random.uniform(bounds[0], bounds[1])
    return individual

# Calculate population diversity
def calculate_diversity(population):
    return np.mean(np.std(population, axis=0))

# Genetic Algorithm
def genetic_algorithm(dimensions=10, bounds=(-10, 10),
                      population_size=50, generations=100, crossover_rate=0.8,
                      initial_mutation_rate=0.2, tournament_size=3, early_stop_generations=10):
    population = initialize_population(population_size, dimensions, bounds)
    best_solution, best_fitness = None, float('inf')
    mutation_rate = initial_mutation_rate
    diversity_history, fitness_history = [], []
    start_time = time.time()

    no_improvement_counter = 0

    for generation in range(generations):
        fitness = evaluate_fitness(population)
        best_idx = np.argmin(fitness)

        if fitness[best_idx] < best_fitness:
            best_solution, best_fitness = population[best_idx].copy(), fitness[best_idx]
            no_improvement_counter = 0
        else:
            no_improvement_counter += 1

        # Early stopping
        if no_improvement_counter >= early_stop_generations:
            print(f"Stopping early at generation {generation} due to no improvement.")
            break

        # Diversity tracking
        diversity = calculate_diversity(population)
        diversity_history.append(diversity)
        fitness_history.append(best_fitness)

        # Dynamic mutation adjustment
        mutation_rate = max(0.01, initial_mutation_rate * (diversity / (bounds[1] - bounds[0])))

        # Selection
        selected_population = tournament_selection(population, fitness, tournament_size)

        # Crossover
        new_population = []
        for i in range(0, len(selected_population), 2):
            parent1, parent2 = selected_population[i], selected_population[(i + 1) % len(selected_population)]
            child1, child2 = uniform_crossover(parent1, parent2, crossover_rate)
            new_population.extend([child1, child2])

        # Mutation
        population = np.array([mutation(ind, mutation_rate, bounds) for ind in new_population])

    # Local refinement of the best solution
    refined_solution = minimize(objective_function, best_solution, bounds=[bounds] * dimensions).x
    refined_fitness = objective_function(refined_solution)

    if refined_fitness < best_fitness:
        best_solution, best_fitness = refined_solution, refined_fitness

    end_time = time.time()
    runtime = end_time - start_time

    # Reporting
    print(f"Best Solution: {best_solution}")
    print(f"Best Fitness: {best_fitness}")
    print(f"Runtime: {runtime:.2f} seconds")



# Run the genetic algorithm
genetic_algorithm()
```
